```python
import pandas as pd
import numpy as np
import scipy as sp
from sklearn.metrics.pairwise import cosine_similarity
import pyarrow as pa
import pyarrow.parquet as pq
```

```python
machineLearningData = pd.read_parquet('data/machineLearning.parquet')
machineLearningDf = pd.DataFrame(machineLearningData)
```

```python
machineLearningDf.head()
```

```python
machineLearningDf = machineLearningDf.apply(lambda x: (x-np.mean(x))/(np.max(x)-np.min(x)), axis=1)
machineLearningDf = machineLearningDf.fillna(0)
machineLearningDf = machineLearningDf.T
machineLearningDf = machineLearningDf.loc[:, (machineLearningDf != 0).any(axis=0)]
machineLearningDf
```

```python
machineLearningSparseDf = sp.sparse.csr_matrix(machineLearningDf.values)
machineLearningSparseDf
```

```python
itemSimilarity = cosine_similarity(machineLearningDf)
```

```python
#item similarity dataframe
itemSimDf = pd.DataFrame(itemSimilarity, index = machineLearningDf.index, columns = machineLearningDf.index)
```

```python
def recommendItem(itemId):
    itemSimData = pd.read_parquet('data/itemSim.parquet')
    itemSimDf = pd.DataFrame(itemSimData)
    counter = 1
    result = {'Aquí hay juegos similares a': itemId,'1':'','2':'','3':'','4':'','5':''}
    for item in itemSimDf.sort_values(by = itemId, ascending = False).index[1:6]:
        result[str(counter)] = item
        counter +=1
    return result
recommendItem('theHunter: Primal')
```

```python
itemSimDf.to_parquet('data/itemSim.parquet')
```
